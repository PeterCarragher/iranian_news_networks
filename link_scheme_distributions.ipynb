```python
import pandas as pd

df_clique = pd.read_csv('data/link_scheme_clique_edges.csv')
df_multi_category = pd.read_csv('data/link_scheme_multi_category_edges.csv')

df_clique_top1k = df_clique.groupby('domain_from').apply(lambda x: x.sort_values('links', ascending=False).head(1000)).reset_index(drop=True)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew
from scipy.stats import entropy

def compute_skewness_entropy(df, n=1000, offset=0):
    # Group the DataFrame by 'domain_from' and calculate the skewness and entropy
    df = df.groupby('domain_from').apply(lambda x: x.sort_values('links', ascending=False).head(n).tail(n-offset)).reset_index(drop=True)
    result = df.groupby('domain_from')['links'].agg([lambda x: skew(x), lambda x: entropy(x)])

    # Rename the columns
    result.columns = ['skewness', 'entropy']

    return result

compute_skewness_entropy(df_clique, 1000)
```

```python
compute_skewness_entropy(df_multi_category, 1000)
```
